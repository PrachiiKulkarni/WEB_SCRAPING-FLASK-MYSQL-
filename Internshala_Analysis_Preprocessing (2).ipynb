```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ast
```

# Importing the scraped dataset

```python
df = pd.read_csv(r"E:\PRACHI\Advance Python\internshala_final_df1.csv")
```

```python
df
```

```python
# Display skills

skills_1=[]
for i in df['Skills_required']:
    skills_1.append(str(i).replace('[','').replace(']','').replace("'",''))
```

```python
# skills_1
```

```python
skills=[]                                 # converting 'skills' column string to list
for i in df['Skills_required']:
    try:
        skills.append(ast.literal_eval(i))
    except:
        k=[np.NaN]
        skills.append(k)
```

```python
# Dropdown Skills - cleaning the dataset for the same meaning skills but different names

skillsRequired=[]
for i in skills:
    k = ['ASP.NET' if j == '.NET' else 'Amazon Web Services (AWS)' if j=='Amazon Web Server (AWS)' else 'MySQL' if j=='SQL'else 'Product Lifecycle Management(PLM)' if j=='Product Management'else j for j in i]
    skillsRequired.append(k)
```

```python
# skillsRequired
```

```python
# Duration                        # cleaning the 'duration' column to numeric values only

Duration_=[]
for i in df['Duration']:
    Duration_.append(i[0])
    
# Duration_
```

```python
# Stipend_given                     # cleaning the 'stipend' column to values only

Stipend_given_=[]
for i in df['Stipend_given']:
    try:
        a = int(i.index('/'))
#         print(a)
        Stipend_given_.append(i[:a-1])
    except:
        Stipend_given_.append(np.NaN)
        
        
# Stipend_given_
```

```python
Location_=[]                               # cleaning the 'location' column for multiple location internships
for i in df['Location']:
    Location_.append(list(i.split(', ')))
        
        
# Location_
```

```python
del df['Duration'] 
del df['Stipend_given']
del df['Skills_required']
```

```python
df
```

```python
df['Location_1'] = Location_    # will be used for Dropdown 'location' in selecting location in 'chooseskills.html'
df['Duration(Months)'] = Duration_
df['Stipend_given(perMonth)'] = Stipend_given_
df['Skills_required'] = skills_1             # will be used for Display 'skills' in final output page
df['Skills_required_1'] = skillsRequired    # will be used for Dropdown 'skills' in selecting skills in 'chooseskills.html'
```

```python
df
```

```python
df.head()
```

```python
skills_1NF = list(map(len, df['Skills_required_1'].values))               # doing 1NF for skills column
# skills_1NF
```

```python
df_1NF = pd.DataFrame({'Internship_title': np.repeat(df['Internship_title'], skills_1NF),
                    'Company': np.repeat(df['Company'], skills_1NF),
                    'Location': np.repeat(df['Location'], skills_1NF),
                    'Location_1': np.repeat(df['Location_1'], skills_1NF),
                    'Duration(Months)': np.repeat(df['Duration(Months)'], skills_1NF),
                    'Stipend_given(perMonth)': np.repeat(df['Stipend_given(perMonth)'], skills_1NF),
                    'Skills_required': np.repeat(df['Skills_required'],skills_1NF),
                    'Skills_required_1': np.concatenate(df['Skills_required_1'].values)})

# df_1NF
```

```python
df_1NF.head()
```

```python
location_1NF = list(map(len, df_1NF['Location_1'].values))     # doing 1NF for location column
# location_1NF
```

```python
df_1NF_final = pd.DataFrame({'Internship_title': np.repeat(df_1NF['Internship_title'], location_1NF),
                    'Company': np.repeat(df_1NF['Company'], location_1NF),
                    'Location': np.repeat(df_1NF['Location'], location_1NF),
                    'Location_1': np.concatenate(df_1NF['Location_1'].values),
                    'Duration(Months)': np.repeat(df_1NF['Duration(Months)'], location_1NF),
                    'Stipend_given(perMonth)': np.repeat(df_1NF['Stipend_given(perMonth)'], location_1NF),
                    'Skills_required': np.repeat(df_1NF['Skills_required'],location_1NF),
                    'Skills_required_1': np.repeat(df_1NF['Skills_required_1'],location_1NF)})

df_1NF_final
```

```python
df_1NF_final.insert(0, 'id', range(1, 1 + len(df_1NF_final)))   # adding 'id' column in datset to be used as primary key in sql
```
