# Importing selenium 

```python
from selenium import webdriver
import time
```

# Looping through 'Internshala' page for extracting data

```python
driver  = webdriver.Chrome('C:\\Webdrivers\\chromedriver.exe') 
driver.get("https://internshala.com/")                                  # opening 'Internshala' website

time.sleep(1)

x1 =driver.find_element_by_xpath('//*[@id="search"]')                   # xpath for search tab on 'internshala' page
x1.click()
time.sleep(1)

x2 =driver.find_element_by_xpath('//*[@id="internship_modal_search_input"]')         #entering values in search tab
x2.send_keys('Data Science')
time.sleep(1)

x3 =driver.find_element_by_xpath('//*[@id="internship_job_search"]/span[2]')         #clicking the search button
x3.click()
time.sleep(1)

x4 =driver.find_element_by_xpath('//*[@id="close_popup"]')                          # closing un-necessary pop-ups
x4.click()
time.sleep(1)

viewbuttons2=[]
text2=[]
href2=[]

for i in range(1,6):                                                                 # looping through page 
    internshiplist=driver.find_elements_by_id('internship_list_container')
    for i in internshiplist:                                                          # to loop through internships only(avoid marketing tabs)
        p1=i.find_elements_by_class_name('container-fluid.individual_internship')
        for j in p1:
            p2=j.find_element_by_class_name('button_container')
            p3=p2.find_element_by_class_name('view_detail_button')
            viewbuttons2.append(p3)
            href2.append(p3.get_property('href'))
            text2.append(j.text)
    
    driver.find_element_by_id('navigation-forward').click()
    time.sleep(1)
```

```python
len(href2)
```

```python
text3=[]
for i in text2:
    text3.append(i.split('\n'))                    # splitting the texts to get further details
```

```python
Heading=[]
Company=[]
Work_type=[]
Duration=[]
Stipend=[]

for i in text3:
    Heading.append(i[0])
    Company.append(i[1])
    Work_type.append(i[2])
    Duration.append(i[6])
    Stipend.append(i[8])
```

```python
dictionary = {'Internship_title': Heading, 'Company': Company, 'Location': Work_type, 'Duration':Duration,'Stipend_given':Stipend, 'href':href2}

df = pd.DataFrame(dictionary)
df                            
```

# Using different driver for getting skills details

# Reading 'href' column from csv file stored earlier 

```python
import pandas as pd
data = pd.read_csv(r'E:\PRACHI\Advance Python\internshala_dataframe1.csv', header=0)
# data
href2 = list(data.href)
# href2
```

# Looping through 'href' list 

```python
details3=[]
for i in href2:
    driver = webdriver.Chrome('C:\\Webdrivers\\chromedriver.exe')
    driver.get(i)                                                  # opening driver page for every href in list
    time.sleep(1)
    
    try:
        x7 = driver.find_element_by_xpath('//*[@id="close_popup"]')         # closing unnecessary pop-ups
        x7.click()              # clicking the page
        time.sleep(1)
    except:
        continue
    
    try: 
        x7 = driver.find_element_by_class_name("internship_details")          # getting the 'Skills-required' if present
        x8 = x7.find_element_by_class_name("round_tabs_container")
        time.sleep(1)
        details3.append(x7.text.split('\n'))
    except:
        details3.append('None')
        continue

    driver.quit()                                                         # quitting the driver to process next href
```

```python
len(details3)                           # got 188 observation correctly
```

```python
skills=[]

for i in range(len(details3)):                      # to get 'Skills' tab only from texts(if present)
    try:
        x=details3[i].index("Skill(s) required")
        y=details3[i].index("Learn these skills on Internshala Trainings")
        skills.append(str(details3[i][x+1:y]))   
    except:
        skills.append(str('Not Found'))
```

```python
len(skills)
```

```python
skills_df = pd.DataFrame(skills)                      # creating dataframe to append to main dataframe
skills_df.columns=['Skills_required']
```
